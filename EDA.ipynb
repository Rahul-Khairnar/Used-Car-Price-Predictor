# EXPLORING THE OWNER TYPE COLUMN

```python
df["Owner_Type"].value_counts()
```

FUNCTION TO CONVERT OWNER TYPES INTO TWO CATEGORIES

```python
def owner_type(type):
    if type == "Second":
        return "Second or More"
    elif type == "Third":
        return "Second or More"
    elif type == "Fourth & Above":
        return "Second or More"
    else:
        return type
```

```python
df["Owner_Type"] = df["Owner_Type"].apply(lambda x:owner_type(x)) 
df["Owner_Type"].value_counts()
```

# EXPLORING THE FUEL TYPE COLUMN

```python
df["Fuel_Type"].value_counts()
```

FUNCTION TO CONVERT LPG CNG AND ELECTRIC TO OTHERS

```python
def fuel_type(type):
    if(type == "CNG") or (type == "LPG") or (type == "Electric"):
        return "Others"
    else:
        return type
```

DELETING OTHERS CATEGORY

```python
df["Fuel_Type"] = df["Fuel_Type"].apply(lambda x:fuel_type(x))
df["Fuel_Type"].value_counts()
df1 = df[~(df["Fuel_Type"] == "Others")]
df1["Fuel_Type"].value_counts()
```
